```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Data/clean_train_data.csv')
df
```

```python
df.columns
```

```python
x=df.drop(['Unnamed: 0', 'Id', "Next Week's Deaths"],axis=1)
y = df["Next Week's Deaths"]
x
```

```python
from flaml import AutoML
automl = AutoML()
```

```python
automl.fit(x, y,task="regression",metric='rmse',time_budget=60*60*1)
```

```python
print('Best ML leaner:', automl.best_estimator)
print('Best hyperparmeter config:', automl.best_config)
print('Best rmse on validation data: {0:.4g}'.format(automl.best_loss))
print('Training duration of best run: {0:.4g} s'.format(automl.best_config_train_time))
```

```python
automl = AutoML(n_estimators = 94, max_features = 0.18217225195335948,max_leaves=17906,FLAML_sample_size=116240)
```

```python
automl.fit(x, y,task="regression",metric='rmse',time_budget=60*60*1)
```

```python
test = pd.read_csv("Data/clean_test_data.csv")
test.drop("Unnamed: 0",axis=1,inplace=True)
test.head()
```

```python
pred = automl.predict(test)
```

```python
pred = automl.predict(test)
test["Next Week's Deaths"] = pred
test.set_index(keys="Id",inplace=True)
```
