```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('Data/train.csv')
test = pd.read_csv('Data/test.csv')
```

```python
train.head()
```

```python
missing_features = train.columns.tolist()
missing_features.remove("Next Week's Deaths")
missing_features.remove("Location")
missing_features.remove("Year")

for feature in missing_features:
    train[feature] = train[feature].fillna(train.groupby(['Location','Year'])[feature].transform(np.nanmean))
    train[feature] = train[feature].fillna(value=0)
    test[feature] = test[feature].fillna(test.groupby(['Location','Year'])[feature].transform(np.nanmean))
    test[feature] = test[feature].fillna(value=0)
    
#
```
